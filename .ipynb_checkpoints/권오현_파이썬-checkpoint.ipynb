# 코로나 19 확진자수 지도 만들기

```python
df = pd.read_csv('./py_data/covid19.csv')
df.head()
```

```python
del df['순번']
```

```python
df.info()
```

```python
df['지역'].unique()
```

```python
a = df['지역'].unique()
for i in a :
    if i == '서울' :
        df_s = df[df['지역'] == i]

df_s
```

```python
df1 = df.groupby('지역').sum()
df1
```

```python
df9 = pd.read_csv('./py_data/각시도별중심위경도.csv', encoding = 'euc-kr')
display(df9.head())
del df9['Unnamed: 0']
del df9['지도center']
df9.head()
```

```python
# 데이터프레임 머지
df5 = pd.merge(df1, df9, how='left', on = '지역')
df5
```

```python
import folium

a = 36.50420   # 위도
b = 127.267826 # 경도
center = [a, b]

m = folium.Map(location = center, zoom_start = 7)
i = 1
for i in df5.index :
    folium.CircleMarker(location = [df5['위도'][i],df5['경도'][i]],
                        radius = round(df5['지역발생수'][i]/400),
                        color='#F19271', 
                        fill_color = 'red',
                        popup = df5['지역'][i]+' '+ str(df5['지역발생수'][i])+ '명').add_to(m)
m
```
