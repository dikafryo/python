## 조건문
- 조건의 결과에 따라 실행되는 문장이 결정
- if, for, while, break, countinue
- 들여쓰기를 지켜야한다.


```python
# if
'''
if 조건 :
    실행문장
elif 조건 :
    실행문장
else :
    실행문장
'''

# 근속년수를 입력받아 근무년수가 20년이상이면, 상여 50만원, 아니면 30만원을 지급

a = int(input('근무년수 : '))

if a >= 20 :
    print('50만원')
else :
    print('30만원')
```

```python
# 나이를 입력받아, 나이가 8세 이하이거나, 65세 이상이면 입장료 무료 출력
# 아니면 입장료 5000원 출력

iAge = int(input('나이 입력 : '))
if iAge >= 65 or iAge <= 8 :
    print('입장료 무료')
else :
    print('5000원')
# & and
# | or
```

```python
iAge = int(input('나이 입력 : '))
if 8 < iAge < 65 :
    print('5000원')
else :
    print('입장료 무료')
```

```python
# 중첩 if
if 날씨가좋다 :
    if 약속 :
        약속 준비
    else :
        놀러간다
else :
    if 약속 :
        약속취소
    else
        잔다
```

# 반복문 (while, for)
- while : 조건문이 참이면 반복, 거짓이면 멈추고 빠져나온다.
- for : 지정확 조건만큼 실행

```python
a = 1
while a <= 10:
    a = a + 1
    if a == 7 :
        break
    if a % 2 == 1 :
        continue
    print(a, '출력')
    
```

```python
# for 변수 in range() :
#     실행문장

# for i in [1,2,3,4] :
#     print(i)

for i in range(5, 11) :
    print(i)
```

```python
# for 문 이용, heelo python 5번 출력.
for i in range(5) :
    print('hello', end = ' ')
    print('python')
```

```python
# 구구단을 입력받아 구구단을 출력
a = int(input('구구단을 선택해주세요 : '))
for i in range(1, 10) :
    print(a , 'x' , i, '=', a*i)
```

```python
for i in range(10) :
    if i == 5 :
        continue
    if i == 8 :
        break
    
    print(i, '출력')
```

```python
'''
for 중첩
*
**
***
...
*********
'''

for i in range(1, 10) :
    print(i, end = ' ')
    for j in range(i) :
        print('*', end = '')
    print()
```

```python
# 영어 학습프로그램 작성
words = {'학교' : 'school', '컴퓨터' : 'computer', '책상' : 'desk'}

for i in words :
    eng = input(i + ' 영어스펠링은? ')
    if eng.upper() == words[i].upper() :
        print('정답')
    else :
        print('오답')
```

# 예외처리 except

```python
'''
try :
    예외처리 유발하는 구문...
except :
    예외처리 실행
else :
    예외없이 실행이 완료된경우에 실행
finally :
    예외 발생여부와 상관없이 실행(항상)
'''
```

```python
try :
    10/5
except Exception as e :
    print('오류정보는 : ' ,str(e))
else :
    print('ok')
finally :
    print('---end---')
```
