```python
# 교재 39 여러장의 이미지를 for로 높이와 너비를 읽어 csv로 저장
import pandas as pd

dic=[]
fileName=['dog (1)','dog (2)','dog (3)','dog (4)','dog (5)']
for i in fileName:
    im=Image.open('./images/' + i + '.jpg')
    dic.append({'파일명':im.filename,'사진높이':im.height,'사진너비':im.width})

df=pd.DataFrame(dic,columns=['파일명','사진높이','사진너비'])
print(df)
df.to_csv('./사진자료정리.csv')
```

```python
# 교재 43 page
# 제어문과 반복문을 이용한 이미지 로고 합성
# 강아지 사진위에는 ‘dog.png’ 
# 고양이 사진위에는 ‘cat.png’를 추가합니다.

import os

path_dir = './images/'
fileList = os.listdir(path_dir)

dog = Image.open(path_dir+'dog.png') # 덮어쓸 이미지 정보
area_dog = (0,0,dog.width,dog.height) # 덮어쓰는 이미지 위치지정
cat = Image.open(path_dir+'cat.png')
area_cat = (0,0,cat.width,cat.height)
for i in fileList:
    gubun = i[:3]    
    im=Image.open(path_dir + i)
    if gubun == 'dog':
        im.paste(dog,area_dog)
    elif gubun == 'cat':
        im.paste(cat,area_cat)
    
    im.save('./img_logo/logo_'+ i)
    
print('---end---')    
```

```python
# 특정 이미지의 정보(사이즈)를 리스트에 넣어서 관리
dog = Image.open('./images/dog.png')
cat = Image.open('./images/cat.png')

area = [[0,0, dog.width, dog.height], [0,0, cat.width, cat.height]]
print(area[0])
print(area[1])
```

```python
# 리스트 정보를 읽어서 짧게 작성 가능
path_dir = './images/'
fileList = os.listdir(path_dir)

for i in fileList:
    gubun = i[:3]    
    im=Image.open(path_dir + i)
    if gubun == 'dog':
        im.paste(dog,area[0])
    elif gubun == 'cat':
        im.paste(cat,area[1])
    
    im.save('./img_logo/logo2_'+ i)

print('---end---')
```
