```python
# 사지선다 문제 만들기
import random

words = [['학교', 'school'], ['컴퓨터' , 'computer'], ['책상' , 'desk'], ['의자', 'chair'],
         ['전화', 'telephone'], ['가족', 'family']]

print('문제를 시작합니다.\n사지선다 문제이므로, 숫자만 입력하세요.')

for i in range(3) :
    choice = random.sample(words, 4) # word 리스트 중에 4개를 가져온다.
    aindex = random.randint(0,3)     # 0~3까지 숫자중 랜덤으로 가져온다.
    
    print(f'{i+1}번. {choice[aindex][1]}의 뜻은?')
    for j in range(4) :
        print(f'{j+1}. {choice[j][0]}')
              
    ans = int(input('정답은 몇번입니까?'))
    if ans == aindex + 1 :
        print('정답')
    else :
        print('오탑')
```

```python
# 사지선다 문제 만들기
import random

words = [['학교', 'school'], ['컴퓨터' , 'computer'], ['책상' , 'desk'], ['의자', 'chair'],
         ['전화', 'telephone'], ['가족', 'family']]

print('문제를 시작합니다.\n사지선다 문제이므로, 숫자만 입력하세요.\n')

user_go = True
i = 1
while user_go :
    choice = random.sample(words, 4) # word 리스트 중에 4개를 가져온다.
    aindex = random.randint(0,3)     # 0~3까지 숫자중 랜덤으로 가져온다.
    
    print(f'{i}번. {choice[aindex][1]}의 뜻은?')
    for j in range(4) :
        print(f'{j+1}. {choice[j][0]}')
              
    uinput = int(input('정답을 입력하세요 : 종료 : 0 >>>>'))
    if uinput == 0 :
        user_go = False
        print('퀴즈 종료')
    elif int(uinput) == aindex + 1 :
        print('정답')
    else :
        print('오탑')
        
    i += 1
    print()
    
```

```python
from tkinter import *
from random import *
from time import sleep

#가장 큰 수
MAX_NUM = 23 

size = 600
window = Tk()
window.title('숫자를 맞춰라!!')

canvas = Canvas(window, width = size, height = size)
canvas.pack()

def nextNumber() :
    canvas.delete('all')

    for i in range(300) :
        col = choice(['pink', 'orange','purple','yellow', 'red', 'blue','coral','indigo','ivory','skyblue'])
        x0 = randint(0, size)
        y0 = randint(0, size)
        d = randint(0, size/5)
        canvas.create_oval(x0,y0,x0+d, y0+d, fill = col)    
        window.update()
        sleep(0.01)

    canvas.create_oval(200,200,400, 400, fill = 'white')
    canvas.create_text(300,300,text=str(randint(1,MAX_NUM)), fill='red', font = ('Helvetica', 60))

btn = Button(window, text="Press Enter", command= nextNumber)
btn.pack(ipadx=10)

window.bind('<Return>',lambda event:callback())
window.mainloop()
```

```python
import pandas as pd
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
filename = "./py_data/driver_quiz.csv"
#pandas read_csv로 불러오기
df = pd.read_csv(filename)
df = df.fillna('')
all_cnt = 0
ok_cnt = 0
while True : ###
    quiz_choice = random.randint(0,len(df)-1)
    print()
    for j in range(5) :
        print(df.columns[j], df.loc[quiz_choice][j])
        if (j == 0) & (df.loc[quiz_choice][6] != '') :
            plt.plot()
            plt.axis("off") #눈금자
            plt.imshow(mpimg.imread('./py_data/images/'+df.loc[quiz_choice][6]))
            plt.show()
    print()
    all_cnt += 1
    user_input = int(input('입력확인. 종료 : 0 >>>>   '))
    if user_input == 0 :   ###        
        print(f'퀴즈 종료합니다. 총 {all_cnt} 문제중 {ok_cnt}문제를 맞췄습니다.')  ###
        break;
    elif user_input == df.loc[quiz_choice][5]:
        ok_cnt += 1
        print('정답입니다.')
    else :
        print(f'오답입니다. 정답은 {df.loc[quiz_choice][5]}입니다.')
    print()
```
